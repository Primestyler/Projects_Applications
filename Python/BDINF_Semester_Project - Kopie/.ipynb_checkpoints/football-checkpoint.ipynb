```python
# Import necessary libraries
import requests
import pandas as pd
import numpy as np


# Define function to fetch data from football-data.org API
def fetch_data(endpoint):
    base_url = "https://api.football-data.org/v4"
    headers = {"X-Auth-Token": "e2ccc8d2448146f5b4576abf56a7b811"}  # Use your personal API token here
    response = requests.get(base_url + endpoint, headers=headers)
    data = response.json()
    return data

# Define function to retrieve statistics for a specific league
def get_league_statistics(league_code):
    standings = fetch_data(f"/competitions/{league_code}/standings")
    
    # Extract relevant statistics
    table = standings["standings"][0]["table"]
    statistics = []
    for team_data in table:
        team_stats = {
            "Team": team_data["team"]["name"],
            "Position": team_data["position"],
            "Played": team_data["playedGames"],
            "Won": team_data["won"],
            "Draw": team_data["draw"],
            "Lost": team_data["lost"],
            "Points": team_data["points"],
            "Goals For": team_data["goalsFor"],
            "Goals Against": team_data["goalsAgainst"],
            "Goal Difference": team_data["goalDifference"]
        }
        statistics.append(team_stats)
    
    return statistics

# Fetch statistics for each league
premier_league_stats = get_league_statistics("PL")
bundesliga_stats = get_league_statistics("BL1")
la_liga_stats = get_league_statistics("PD")

# Create DataFrames
premier_league_df = pd.DataFrame(premier_league_stats)
bundesliga_df = pd.DataFrame(bundesliga_stats)
la_liga_df = pd.DataFrame(la_liga_stats)

# Display the DataFrames
print("Premier League Statistics:")
print(premier_league_df)

print("\nGerman Bundesliga Statistics:")
print(bundesliga_df)

print("\nSpanish La Liga Statistics:")
print(la_liga_df)
```

```python
# Summary statistics
print("Summary Statistics for Premier League:")
print(premier_league_df.describe())

# Visualize key metrics
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

# Goals scored distribution
plt.subplot(2, 2, 1)
plt.hist(premier_league_df["Goals For"], bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution of Goals Scored")
plt.xlabel("Goals")
plt.ylabel("Frequency")

# Points earned distribution
plt.subplot(2, 2, 2)
plt.hist(premier_league_df["Points"], bins=20, color="salmon", edgecolor="black")
plt.title("Distribution of Points Earned")
plt.xlabel("Points")
plt.ylabel("Frequency")

# Remove non-numeric columns for correlation matrix
numeric_columns = premier_league_df.select_dtypes(include=[np.number]).columns
premier_league_numeric = premier_league_df[numeric_columns]

# Correlation matrix
plt.subplot(2, 2, 3)
correlation_matrix = premier_league_numeric.corr()
plt.imshow(correlation_matrix, cmap="coolwarm", interpolation="nearest")
plt.colorbar()
plt.title("Correlation Matrix")
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

plt.tight_layout()
plt.show()

```

Distribution of Goals: This graph illustrates the distribution of goals in the Premier League. You can observe that the distribution is fairly symmetrical, with most teams scoring a moderate number of goals. Some teams score a significant number of goals, while others score fewer. This may indicate various tactical and performance differences between the teams.

Distribution of Points: This graph displays the distribution of points in the Premier League. You can see that the distribution is rather right-skewed, with most teams earning a moderate number of points. However, a few teams achieve a high score, while others earn only a few points. This reflects the performance disparities between the teams and identifies which teams are more successful than others.

Correlation Matrix: This matrix exhibits the correlation between various numerical variables in your data. A value close to 1 or -1 indicates a strong positive or negative correlation, while a value close to 0 suggests a weak or no correlation. Analyzing this matrix allows you to determine which variables in your data are related to each other. For instance, there could be a strong positive correlation between goals scored and points earned, implying that teams scoring more goals tend to accumulate more points.


```python
# Summary statistics for Bundesliga
print("\nSummary Statistics for German Bundesliga:")
print(bundesliga_df.describe())

# Summary statistics for La Liga
print("\nSummary Statistics for Spanish La Liga:")
print(la_liga_df.describe())

# Visualize key metrics for Bundesliga
plt.figure(figsize=(12, 8))

# Goals scored distribution
plt.subplot(2, 2, 1)
plt.hist(bundesliga_df["Goals For"], bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution of Goals Scored (Bundesliga)")
plt.xlabel("Goals")
plt.ylabel("Frequency")

# Points earned distribution
plt.subplot(2, 2, 2)
plt.hist(bundesliga_df["Points"], bins=20, color="salmon", edgecolor="black")
plt.title("Distribution of Points Earned (Bundesliga)")
plt.xlabel("Points")
plt.ylabel("Frequency")

# Remove non-numeric columns for correlation matrix
numeric_columns_bundesliga = bundesliga_df.select_dtypes(include=[np.number]).columns
bundesliga_numeric = bundesliga_df[numeric_columns_bundesliga]

# Correlation matrix for Bundesliga
plt.subplot(2, 2, 3)
correlation_matrix_bundesliga = bundesliga_numeric.corr()
plt.imshow(correlation_matrix_bundesliga, cmap="coolwarm", interpolation="nearest")
plt.colorbar()
plt.title("Correlation Matrix (Bundesliga)")
plt.xticks(range(len(correlation_matrix_bundesliga.columns)), correlation_matrix_bundesliga.columns, rotation=90)
plt.yticks(range(len(correlation_matrix_bundesliga.columns)), correlation_matrix_bundesliga.columns)

plt.tight_layout()
plt.show()

# Visualize key metrics for La Liga
plt.figure(figsize=(12, 8))

# Goals scored distribution
plt.subplot(2, 2, 1)
plt.hist(la_liga_df["Goals For"], bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution of Goals Scored (La Liga)")
plt.xlabel("Goals")
plt.ylabel("Frequency")

# Points earned distribution
plt.subplot(2, 2, 2)
plt.hist(la_liga_df["Points"], bins=20, color="salmon", edgecolor="black")
plt.title("Distribution of Points Earned (La Liga)")
plt.xlabel("Points")
plt.ylabel("Frequency")

# Remove non-numeric columns for correlation matrix
numeric_columns_la_liga = la_liga_df.select_dtypes(include=[np.number]).columns
la_liga_numeric = la_liga_df[numeric_columns_la_liga]

# Correlation matrix for La Liga
plt.subplot(2, 2, 3)
correlation_matrix_la_liga = la_liga_numeric.corr()
plt.imshow(correlation_matrix_la_liga, cmap="coolwarm", interpolation="nearest")
plt.colorbar()
plt.title("Correlation Matrix (La Liga)")
plt.xticks(range(len(correlation_matrix_la_liga.columns)), correlation_matrix_la_liga.columns, rotation=90)
plt.yticks(range(len(correlation_matrix_la_liga.columns)), correlation_matrix_la_liga.columns)

plt.tight_layout()
plt.show()
```

Die Spalte "Played" in der Korrelationsmatrix ist komplett weiß, weil sie keine Variation in den Werten aufweist.
In der Premier League haben alle Teams entweder 30 oder 31 Spiele gespielt, während in der Bundesliga und La Liga die Anzahl der gespielten Spiele bei allen Teams mit 28 bzw. 30 gleich ist. 
Wenn eine Spalte in der Korrelationsmatrix keine Variation aufweist, bedeutet dies, dass es keinen Unterschied zwischen den Werten gibt, und daher gibt es keine Korrelation zwischen dieser Spalte und anderen numerischen Variablen.

```python
import pymongo

# Verbinde dich mit dem MongoDB-Server
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Überprüfe die Liste der verfügbaren Datenbanken
print(client.list_database_names())
```

```python
# Daten einfügen für die Premier League
db = client["football_data"]
premier_league_collection = db["premier_league"]
premier_league_collection.insert_many(premier_league_stats)

# Daten einfügen für die Bundesliga
bundesliga_collection = db["bundesliga"]
bundesliga_collection.insert_many(bundesliga_stats)

# Daten einfügen für La Liga
la_liga_collection = db["la_liga"]
la_liga_collection.insert_many(la_liga_stats)
```

```python
# Daten aus der Premier League abrufen
premier_league_data = list(premier_league_collection.find())
print("Premier League Data:", premier_league_data)

# Daten aus der Bundesliga abrufen
bundesliga_data = list(bundesliga_collection.find())
print("Bundesliga Data:", bundesliga_data)

# Daten aus La Liga abrufen
la_liga_data = list(la_liga_collection.find())
print("La Liga Data:", la_liga_data)
```
