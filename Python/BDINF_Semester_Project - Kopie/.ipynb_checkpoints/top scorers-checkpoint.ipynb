```python
# Importieren der benötigten Bibliotheken
import requests
import matplotlib.pyplot as plt
```

```python
# Definieren der Basis-URL der API und des API-Tokens
base_url = "http://api.football-data.org/v4"
headers = {"X-Auth-Token": "e2ccc8d2448146f5b4576abf56a7b811"}
```

```python
# Funktion zur Abrufung von Daten über die football-data.org API
def fetch_data(endpoint):
    response = requests.get(base_url + endpoint, headers=headers)
    data = response.json()
    return data
```

```python
# Funktion zur Extraktion von Daten über die Top-Torschützen eines Wettbewerbs
def extract_top_scorers_data(competition_id, num_players):
    scorers_data = fetch_data(f"/competitions/{competition_id}/scorers")
    top_scorers = scorers_data["scorers"][:num_players]  # Die ersten num_players Top-Torschützen
    return top_scorers
```

```python
# calculate current age of player derived from dateOfBirth
from datetime import datetime
import math

print("datetime now: ", datetime.now())
end_date = datetime.now()
top_scorer_birthday = extract_top_scorers_data(2019, 5)[0]["player"]["dateOfBirth"]
top_scorer_birthday_datetime = datetime.strptime(top_scorer_birthday, '%Y-%m-%d')
print("player's birthday: ", top_scorer_birthday_datetime)
start_date = top_scorer_birthday_datetime
difference_in_days = (end_date - start_date).days
difference_in_years = math.floor(difference_in_days / 365)
print("--> age: ", difference_in_years)
```

```python
# checking types of list elements
print(type(top_scorers))
print(type(top_scorers[0]))
print(type(top_scorers[0]["player"]))

# add field "age" to players
today_datetime = datetime.today()

for x in top_scorers:
  scorer_birthday_datetime = datetime.strptime(x["player"]["dateOfBirth"], '%Y-%m-%d')
  x["player"]["age"] = math.floor((today_datetime - scorer_birthday_datetime).days / 365)
  print(x["player"], "| goals: ", x["goals"])
```

```python
# scatterplot of goals over player's age

#print(len(top_scorers))
x = [] 
y = [] 

for t in top_scorers:
    x.append(t.get("player").get("age"))
    y.append(t.get("goals"))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y)
ax.set_xlabel('age')
ax.set_ylabel('goals')
ax.set_title('Scatter plot of goals over age')

plt.show()
```

```python
# Funktion zur Extraktion von historischen Daten über die Top-Torschützen
def extract_historical_top_scorers_data(competition_id, season_start, season_end, num_players):
    historical_data = []
    for season in range(season_start, season_end + 1):
        scorers_data = fetch_data(f"/competitions/{competition_id}/scorers?season={season}")
        if "scorers" in scorers_data:
            top_scorers = scorers_data["scorers"][:num_players]
            historical_data.append({"season": season, "top_scorers": top_scorers})
        else:
            print(f"Keine Daten für die Saison {season} gefunden.")
    return historical_data
```

```python
# Funktion zur Visualisierung der prozentualen Verteilung der Tore jedes Top-Torschützen
def plot_goal_distribution(top_scorers):
    player_names = [scorer["player"]["name"] for scorer in top_scorers]
    goals = [scorer["goals"] for scorer in top_scorers]
    plt.figure(figsize=(10, 6))
    plt.pie(goals, labels=player_names, autopct="%1.1f%%")
    plt.title("Prozentuale Verteilung der Tore jedes Top-Torschützen")
    plt.show()

    
```

```python
# Funktion zur Visualisierung der Entwicklung der Anzahl der Tore über verschiedene Saisons hinweg
def plot_goals_over_seasons(historical_data):
    seasons = [data["season"] for data in historical_data]
    max_goals_count = max(len(data["top_scorers"]) for data in historical_data)
    
    plt.figure(figsize=(10, 6))
    for index in range(max_goals_count):
        goals = [data["top_scorers"][index]["goals"] if index < len(data["top_scorers"]) else 0 for data in historical_data]
        plt.plot(seasons, goals, label=f"Top-Torschützen {index + 1}")
    
    plt.xlabel("Saison")
    plt.ylabel("Anzahl der Tore")
    plt.title("Entwicklung der Anzahl der Tore über verschiedene Saisons hinweg")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Liste von Wettbewerben definieren
competitions = [
    {"id": 2019, "name": "Serie A"},
    {"id": 2002, "name": "Bundesliga"},  
    {"id": 2014, "name": "La Liga"} 
]

# Anzahl der Top-Torschützen, die angezeigt werden sollen
num_top_scorers = 5
```

```python
competitions[0]["id"]
```

```python
# Daten für mehrere Wettbewerbe abrufen und anzeigen
for competition in competitions:
    print(f"Top-Torschützen für {competition['name']}:")
    top_scorers = extract_top_scorers_data(competition["id"], num_top_scorers)
    for index, scorer in enumerate(top_scorers, start=1):
        print(f"{index}. {scorer['player']['name']} - Tore: {scorer['goals']}, Vorlagen: {scorer['assists']}, Strafstöße: {scorer['penalties']}")
    print()
    plot_goal_distribution(top_scorers)
```

```python
# Historische Daten für Top-Torschützen abrufen und anzeigen
season_start = 2018  # Startjahr der historischen Daten
season_end = 2022  # Endjahr der historischen Daten
for competition in competitions:
    print(f"Historische Top-Torschützen für {competition['name']}:")
    historical_data = extract_historical_top_scorers_data(competition["id"], season_start, season_end, num_top_scorers)
    for season_data in historical_data:
        season = season_data["season"]
        top_scorers = season_data["top_scorers"]
        print(f"Saison {season}:")
        for index, scorer in enumerate(top_scorers, start=1):
            print(f"{index}. {scorer['player']['name']} - Tore: {scorer['goals']}, Vorlagen: {scorer['assists']}, Strafstöße: {scorer['penalties']}")
        print()
    plot_goals_over_seasons(historical_data)
```

```python
import pymongo

# Verbindung zur lokalen MongoDB-Datenbank herstellen
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Datenbank auswählen oder erstellen
db = client["football_players"]

# Sammlung (Collection) auswählen
collection = db["players"]
```

```python
# Spielerdaten aus der MongoDB abrufen
players_data = collection.find()

# Spielerdaten in Listen speichern
player_names = []
player_goals = []
player_assists = []

for player in players_data:
    player_names.append(player['name'])
    player_goals.append(player['goals'])
    player_assists.append(player['assists'])

# Visualisierung der Daten
plt.figure(figsize=(10, 6))
plt.bar(player_names, player_goals, color='skyblue', label='Tore')
plt.bar(player_names, player_assists, color='orange', label='Vorlagen')
plt.xlabel('Spieler')
plt.ylabel('Anzahl')
plt.title('Tore und Vorlagen der Spieler in der Saison 2023/24')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Spielerdaten nach Nationalität gruppieren und die Gesamtzahl der Tore und Vorlagen berechnen
pipeline = [
    {"$group": {"_id": "$nationality", "total_goals": {"$sum": "$goals"}, "total_assists": {"$sum": "$assists"}}}
]
nationality_stats = list(collection.aggregate(pipeline))

# Visualisierung der Spielerdaten nach Nationalität
nationalities = [stat["_id"] for stat in nationality_stats]
goals = [stat["total_goals"] for stat in nationality_stats]
assists = [stat["total_assists"] for stat in nationality_stats]

plt.figure(figsize=(12, 6))
plt.bar(nationalities, goals, color='skyblue', label='Tore')
plt.bar(nationalities, assists, color='orange', label='Vorlagen', bottom=goals)
plt.xlabel('Nationalität')
plt.ylabel('Anzahl')
plt.title('Tore und Vorlagen nach Nationalität')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
