### Wahl von MongoDB für das Projekt

Für das vorliegende Projekt der Verwaltung umfangreicher Fußballdaten wurde MongoDB als die geeignete Datenbank gewählt. Hier sind die Gründe für diese Entscheidung:

1. **Flexible Datenstruktur**: MongoDB bietet ein flexibles Datenmodell, das sich leicht an verändernde Anforderungen anpassen lässt. In einem Fußballdatenprojekt können die Datenstrukturen je nach Art der Informationen, die gespeichert werden müssen, variieren. MongoDB ermöglicht die Speicherung von unstrukturierten Daten sowie von Daten mit komplexen Hierarchien, was ideal ist, um verschiedene Arten von Fußballdaten effizient zu verwalten.

2. **Skalierbarkeit**: MongoDB ist eine gute Wahl für Projekte, die eine horizontale Skalierung erfordern. In dem Fußballdatenprojekt, insbesondere wenn es um umfangreiche Datensätze wie Spielerinformationen, Spielstatistiken, Ligaergebnisse usw. geht, ist die Fähigkeit, mit wachsenden Datenmengen umzugehen, entscheidend. MongoDB bietet die Möglichkeit, leicht horizontal zu skalieren, indem zusätzliche Server hinzugefügt werden, um die Last zu verteilen.

3. **Leistung**: MongoDB ist für seine Leistungsfähigkeit bekannt, insbesondere wenn es um das Speichern und Abfragen großer Datenmengen geht. In dem Fußballdatenprojekt, in dem eine große Anzahl von Abfragen ausgeführt wird, um Informationen über Spiele, Spielerstatistiken, Tabellenstände usw. zu erhalten, ist eine Datenbank mit hoher Leistung unerlässlich.

4. **Unterstützung für komplexe Abfragen**: MongoDB bietet eine leistungsstarke Abfragesprache sowie die Möglichkeit, komplexe Abfragen mit Aggregationen, Indizes und anderen fortgeschrittenen Funktionen durchzuführen. Dies ist besonders wichtig, um tiefe Einblicke in die Fußballdaten zu gewinnen und komplexe Analysen durchzuführen, wie beispielsweise die Berechnung von Durchschnittswerten, das Erstellen von Diagrammen und vieles mehr.

5. **Integration mit Python**: MongoDB lässt sich nahtlos in Python-Anwendungen integrieren, was für das vorliegende Projekt von Vorteil ist, da der vorhandene Code in Python geschrieben ist. Die offizielle MongoDB Python-Driver-Bibliothek erleichtert die Verbindung zur Datenbank, das Ausführen von Abfragen und das Verarbeiten von Daten in Python-Anwendungen.

Insgesamt bietet MongoDB eine ideale Datenbank für die effiziente Verwaltung und Analyse von Fußballdaten, indem es Flexibilität, Skalierbarkeit, Leistung und umfangreiche Abfragefunktionen kombiniert.


# Theoretische Berücksichtigung der Big Data Kriterien

Unser Projekt zur Verwaltung von Fußballdaten könnte theoretisch in gewisser Weise die Big Data Kriterien berücksichtigen:

## 5 Vs von Big Data:
- **Volumen:** Wir verarbeiten und speichern eine große Menge an Daten, einschließlich umfangreicher Spielerstatistiken aus verschiedenen Fußballligen. Obwohl unser Volumen im Vergleich zu großen Big-Data-Systemen klein ist, ist es dennoch wichtig, die Skalierbarkeit und Handhabung größerer Datenmengen zu berücksichtigen, um für zukünftiges Wachstum gerüstet zu sein.

- **Variety:** Unsere Daten stammen aus verschiedenen Quellen und umfassen unterschiedliche Arten von Informationen, einschließlich Spielerdetails, Teamstatistiken und Spielberichte. Diese Vielfalt erfordert eine flexible Datenmodellierung und -verarbeitung, um sie effektiv zu nutzen.

- **Velocity:** Die Aktualisierungsgeschwindigkeit unserer Daten kann variieren, da neue Spielerstatistiken kontinuierlich verfügbar werden, während Spielberichte weniger häufig aktualisiert werden. Es ist wichtig, diese Geschwindigkeit zu berücksichtigen, um Echtzeit- oder zeitnahe Analysen durchzuführen.

- **Veracity:** Die Qualität unserer Daten ist entscheidend für die Genauigkeit unserer Analysen und Prognosen. Wir müssen sicherstellen, dass die Daten aus zuverlässigen Quellen stammen und korrekt verarbeitet werden, um fehlerhafte Ergebnisse zu vermeiden.

- **Value:** Letztendlich streben wir danach, aus unseren Daten wertvolle Erkenntnisse zu gewinnen, sei es zur Leistungsanalyse von Spielern und Teams. Die Fähigkeit, diese Erkenntnisse in konkrete Aktionen umzusetzen, bestimmt den Wert unseres Projekts.

```python
# Funktion zum Speichern von JSON-Daten in MongoDB
def save_json_to_mongodb(file_path, collection_name):
    # Daten aus der JSON-Datei lesen
    with open(file_path, 'r', encoding='utf-8') as file:
        json_data = json.load(file)
    
    # Überprüfen, ob die Daten nicht leer sind
    if json_data:
        # Wenn die JSON-Daten ein einzelnes Objekt sind, wandeln wir es in eine Liste um
        if not isinstance(json_data, list):
            json_data = [json_data]
        
        # Daten in die MongoDB speichern
        collection = db[collection_name]
        collection.insert_many(json_data)
        print("Daten erfolgreich in die MongoDB gespeichert.")
    else:
        print("Die JSON-Datei enthält keine Daten oder ist ungültig.")

# Dateipfad zu den JSON-Dateien
laliga_file_path = 'laliga_players_2023.json'
premier_league_file_path = 'premier_league_players_2023.json'
bundesliga_file_path = 'bundesliga_players_2023.json'



# Kollektionsnamen für die MongoDB
laliga_collection_name = 'laliga_players'
premier_league_collection_name = 'premier_league_players'
bundesliga_collection_name = 'bundesliga_players'
```

## Analyse der Entwicklung der Top 5 Teams der Premier League (2020-2023)

In diesem Abschnitt wird die Entwicklung der Top 5 Teams der Premier League aus der Saison 2020 über die nächsten 3 Jahre analysiert. Die Daten stammen aus einer MongoDB-Datenbank, die Informationen zu den Tabellenständen der Premier League für verschiedene Spielzeiten enthält.

```python


# Daten für die Jahre 2020 bis 2023 abrufen
top_teams_2020 = db['pl_standings_2020'].find_one()['standings'][0]['table'][:5]

# Teams für die Jahre 2021, 2022 und 2023 abrufen und ihre Entwicklung verfolgen
years = ['2020', '2021', '2022', '2023']
teams_points = {team['team']['name']: [] for team in top_teams_2020}

# Punkte für das Jahr 2020 hinzufügen
for team in top_teams_2020:
    teams_points[team['team']['name']].append(team['points'])

for year in years[1:]:
    standings = db[f'pl_standings_{year}'].find_one()['standings'][0]['table']
    for team_2020 in top_teams_2020:
        for team in standings:
            if team['team']['name'] == team_2020['team']['name']:
                teams_points[team['team']['name']].append(team['points'])
                break
        else:
            teams_points[team_2020['team']['name']].append(0)  # Füge 0 Punkte hinzu, falls das Team nicht gefunden wurde

# Liniendiagramm erstellen
plt.figure(figsize=(10, 6))

for team, points in teams_points.items():
    plt.plot(years, points, marker='o', label=team)

plt.title('Entwicklung der Punkte der Top 5 Teams der Premier League (2020-2023)')
plt.xlabel('Jahr')
plt.ylabel('Punkte')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Diagramm anzeigen
plt.show()
```

Aus dem Liniendiagramm lässt sich ablesen, dass Manchester City seit 2020 konstant die Spitze bildet und somit die letzten Jahre immer Meister wurde. Desweiteren sieht man den drastischen Abstieg von Leicester City, die einst im Jahr 2020 mit 66 Punkten Platz 5 in der Premiere League waren und deren Punkte im Jahr 2023 bei 0 stehen, da sie in der Saison davor abgestiegen sind und nicht mehr an der Premiere League teilnehmen.

### Analyse der Durchschnittlichen Tore pro Spiel (2020-2023) in der Premier League

In diesem Abschnitt werden die durchschnittliche Anzahl der Tore pro Spiel für die Saisons 2020 bis 2023 in der Premiere League analysiert.

```python
# Funktion zum Abrufen der Daten aus der MongoDB für bestimmte Jahre
def get_average_goals_per_game_for_years(collection_names):
    average_goals_per_game = []
    for collection_name in collection_names:
        document = db[collection_name].find_one({})
        year = int(document['season']['startDate'][:4])
        total_goals_for = sum([team['goalsFor'] for team in document['standings'][0]['table']])
        total_goals_against = sum([team['goalsAgainst'] for team in document['standings'][0]['table']])
        total_goals = total_goals_for + total_goals_against  # Berücksichtige sowohl erzielte als auch kassierte Tore
        total_games = sum([team['playedGames'] for team in document['standings'][0]['table']])
        average_goals = total_goals / total_games
        average_goals_per_game.append({"year": year, "average_goals": average_goals})
    return average_goals_per_game

# Daten für die durchschnittliche Anzahl der Tore pro Spiel aus der MongoDB abrufen
collection_names = ['pl_standings_2020', 'pl_standings_2021', 'pl_standings_2022', 'pl_standings_2023']
average_goals_data = get_average_goals_per_game_for_years(collection_names)

# Jahre und durchschnittliche Tore extrahieren
years = [data["year"] for data in average_goals_data]
average_goals = [data["average_goals"] for data in average_goals_data]

# Diagramm erstellen
plt.figure(figsize=(10, 6))
plt.plot(years, average_goals, marker='o')
plt.xlabel('Jahr')
plt.ylabel('Durchschnittliche Tore pro Spiel')
plt.title('Durchschnittliche Tore pro Spiel für die Saisons 2020-2023')
plt.xticks(years)
plt.grid(True)
plt.show()
```

Aus dem Liniendiagramm lässt sich unschwer erkennen, dass seit dem Jahr 2023 ein drastischer Anstieg in der Anzahl der durchschnittliche Tore pro Spiel zu beobachten ist. Dies lässt sich daraus erklären, dass aufgrund eines neuen Reglements, das vorsieht, dass Schiedsrichter eine deutlich längere Nachspielzeit erlauben. Dies hat zur Folge, dass die Partien länger dauern. Und dass ein längeres Spiel zu mehr Torchancen führt, ist ein wichtiger Faktor für die hohe Torausbeute in der Saison 2023.

## Durchschnittsalter, Körpergröße und Gewicht

Nachdem wir die Daten erfolgreich importiert haben, berechnen wir das Durchschnittsalter, die durchschnittliche Körpergröße und das durchschnittliche Gewicht der Spieler in verschiedenen Fußballligen. Durch Visualisierungen stellen wir die Unterschiede zwischen den Ligen anschaulich dar und erkennen mögliche Trends.



Zunächst definieren wir eine neue Funktion unter dem namen "calculate_average_age", mit der wir das Durchschnittsalter aller Spieler einer Liga herausfinden können. Unser Ergebnis zeigt uns, dass das Durschnittsalter in der Bundesliga am niedrigsten ist und in der Premier League mit deutlichem Unterschied am höchsten. 

```python
# Funktion zum Berechnen des Durchschnittsalters der Spieler in einer Liga
def calculate_average_age(collection_name):
    collection = db[collection_name]
    pipeline = [
        {"$unwind": "$response"},  # Array entfalten
        {"$group": {"_id": None, "average_age": {"$avg": {"$toInt": "$response.player.age"}}}}
    ]
    result = list(collection.aggregate(pipeline))
    if result:
        return result[0]["average_age"]
    else:
        return None


# Durchschnittsalter für jede Liga berechnen
laliga_avg_age = calculate_average_age('laliga_players')
premier_league_avg_age = calculate_average_age('premier_league_players')
bundesliga_avg_age = calculate_average_age('bundesliga_players')

# Drucke die Durchschnittsalter
print("Durchschnittsalter in La Liga:", laliga_avg_age)
print("Durchschnittsalter in Premier League:", premier_league_avg_age)
print("Durchschnittsalter in Bundesliga:", bundesliga_avg_age)
```

## Berechnung des Body-Mass-Index (BMI)

Basierend auf den durchschnittlichen Körpergrößen und -gewichten der Spieler berechnen wir auch den durchschnittlichen Body-Mass-Index (BMI). Dies hilft uns, das Gewicht im Verhältnis zur Körpergröße zu betrachten und mögliche Differenzen zu erkennen.



```python
# Funktion zum Berechnen des BMI
def calculate_bmi(height, weight):
    # Umrechnung der Körpergröße von cm in Meter
    height_m = height / 100
    # Berechnung des BMI: Gewicht (kg) / (Körpergröße (m))^2
    bmi = weight / (height_m ** 2)
    return bmi

# Durchschnittliche Körpergröße und Gewicht für jede Liga
laliga_avg_height = calculate_average_height('laliga_players')
laliga_avg_weight = calculate_average_weight('laliga_players')

premier_league_avg_height = calculate_average_height('premier_league_players')
premier_league_avg_weight = calculate_average_weight('premier_league_players')

bundesliga_avg_height = calculate_average_height('bundesliga_players')
bundesliga_avg_weight = calculate_average_weight('bundesliga_players')

# Berechnung des BMI für jede Liga
laliga_avg_bmi = calculate_bmi(laliga_avg_height, laliga_avg_weight)
premier_league_avg_bmi = calculate_bmi(premier_league_avg_height, premier_league_avg_weight)
bundesliga_avg_bmi = calculate_bmi(bundesliga_avg_height, bundesliga_avg_weight)

# Drucke das BMI
print("Durchschnittliches BMI in La Liga:", laliga_avg_bmi)
print("Durchschnittliches BMI in Premier League:", premier_league_avg_bmi)
print("Durchschnittliches BMI in Bundesliga:", bundesliga_avg_bmi)

# Visualisierung des BMI für Spieler in verschiedenen Ligen
leagues = ['La Liga', 'Premier League', 'Bundesliga']
average_bmis = [laliga_avg_bmi, premier_league_avg_bmi, bundesliga_avg_bmi]

# Visualisierung erstellen
plt.figure(figsize=(10, 6))
plt.bar(leagues, average_bmis, color=['blue', 'red', 'green'])
plt.xlabel('Liga')
plt.ylabel('Durchschnittlicher BMI')
plt.title('Durchschnittlicher BMI der Spieler in verschiedenen Ligen')
plt.show()
```

```python
# Funktion zur Berechnung der durchschnittlichen Anzahl der Tore pro Spiel für jedes Team
def calculate_avg_goals_per_team(collection_name):
    collection = db[collection_name]
    cursor = collection.find()

    goals_per_team = {}  
    matches_per_team = {}  

    for document in cursor:
        matches = document['matches']
        for match in matches:
            home_team = match['homeTeam']['name']
            away_team = match['awayTeam']['name']
            home_goals = match['score']['fullTime']['home']
            away_goals = match['score']['fullTime']['away']

            # Überprüfen, ob die Tore nicht None sind, bevor sie hinzugefügt werden
            if home_goals is not None:
                if home_team in goals_per_team:
                    goals_per_team[home_team] += home_goals
                    matches_per_team[home_team] += 1
                else:
                    goals_per_team[home_team] = home_goals
                    matches_per_team[home_team] = 1

            if away_goals is not None:
                if away_team in goals_per_team:
                    goals_per_team[away_team] += away_goals
                    matches_per_team[away_team] += 1
                else:
                    goals_per_team[away_team] = away_goals
                    matches_per_team[away_team] = 1

    avg_goals_per_team = {}  

    for team, goals in goals_per_team.items():
        matches = matches_per_team[team]
        avg_goals_per_team[team] = goals / matches if matches > 0 else 0

    return avg_goals_per_team

    # Durchschnittliche Tore pro Team berechnen
    for team, goals in goals_per_team.items():
        matches = matches_per_team[team]
        avg_goals_per_team[team] = goals / matches if matches > 0 else 0

    return avg_goals_per_team


# Ausgabe der Top 5 Teams
print("Top 5 Teams mit den höchsten durchschnittlichen Toren pro Spiel:")
for team, avg_goals in top_5_teams:
    print(f"{team}: {avg_goals:.2f} Tore pro Spiel")
    import matplotlib.pyplot as plt

# Funktion zum Abrufen der Top 5 Teams mit den höchsten durchschnittlichen Toren pro Spiel
def get_top_5_teams(avg_goals_per_team):
    # Sortiere die Teams nach durchschnittlichen Toren pro Spiel in absteigender Reihenfolge
    sorted_teams = sorted(avg_goals_per_team.items(), key=lambda x: x[1], reverse=True)
    # Wähle die Top 5 Teams aus
    top_5_teams = sorted_teams[:5]
    return top_5_teams

# Durchschnittliche Anzahl der Tore pro Spiel für die UEFA Champions League berechnen
avg_goals_per_team = calculate_avg_goals_per_team('cl_matches')
# Die Top 5 Teams mit den höchsten durchschnittlichen Toren pro Spiel abrufen
top_5_teams = get_top_5_teams(avg_goals_per_team)

# Daten für das Diagramm vorbereiten
teams = [team[0] for team in top_5_teams]
avg_goals = [team[1] for team in top_5_teams]

# Balkendiagramm erstellen
plt.figure(figsize=(10, 6))
plt.barh(teams, avg_goals, color='skyblue')
plt.xlabel('Durchschnittliche Anzahl der Tore pro Spiel')
plt.ylabel('Team')
plt.title('Top 5 Teams mit den höchsten durchschnittlichen Toren pro Spiel (UEFA Champions League)')
plt.gca().invert_yaxis()  # Umkehrung der y-Achse, damit das Team mit den meisten Toren oben ist
plt.show()

```

Mithilfe der Funktion "calculate_avg_goals_home_away" können wir nun die Anzahl an Toren ausrechnen die ein Team im Heimspiel, als auch im Auswärtsspiel gemacht hat und geben daraufhin die Werte für jeden Club aus.

```python
# Funktion zur Berechnung der durchschnittlichen Anzahl von Toren für Heim- und Auswärtsspiele
def calculate_avg_goals_home_away(collection_name):
    collection = db[collection_name]
    pipeline = [
        {"$unwind": "$matches"},
        {"$group": {
            "_id": "$matches.homeTeam.name",
            "home_goals": {"$avg": "$matches.score.fullTime.home"},
            "away_goals": {"$avg": "$matches.score.fullTime.away"}
        }}
    ]
    result = list(collection.aggregate(pipeline))
    return result

# Durchschnittliche Anzahl von Toren für Heim- und Auswärtsspiele für jedes Team berechnen
avg_goals_home_away = calculate_avg_goals_home_away('cl_matches')

# Ergebnis ausgeben
for team_stats in avg_goals_home_away:
    print("Team:", team_stats["_id"])
    print("Durchschnittliche Anzahl von Toren zu Hause:", team_stats["home_goals"])
    print("Durchschnittliche Anzahl von Toren auswärts:", team_stats["away_goals"])
    print()
```

Diese geben wir nun wieder aus. Es ist auffallend, dass die durchschnittlichen Tore in Heimspielen höher ist, als in den Auswärtsspielen. Wir schauen uns dies gleich noch in Diagrammen an.

```python
# Funktion zur Ermittlung der Top 5 Teams für Heim- und Auswärtsspiele
def get_top_5_teams_home_away(avg_goals_home_away, home=True):
    # Filtern der Teams mit gültigen Werten für Heim- oder Auswärtstore
    valid_teams = [team for team in avg_goals_home_away if team["home_goals" if home else "away_goals"] is not None]
    
    # Sortieren der Teams nach den durchschnittlichen Toren
    sorted_teams = sorted(valid_teams, key=lambda x: x["home_goals" if home else "away_goals"], reverse=True)
    
    # Die ersten fünf Teams auswählen
    top_5_teams = sorted_teams[:5]
    return top_5_teams

# Top 5 Teams für Heimspiele
top_5_teams_home = get_top_5_teams_home_away(avg_goals_home_away, home=True)

print("Top 5 Teams für Heimspiele:")
for idx, team_stats in enumerate(top_5_teams_home, start=1):
    print(f"{idx}. Team: {team_stats['_id']}, Durchschnittliche Tore: {team_stats['home_goals']:.2f}")

# Top 5 Teams für Auswärtsspiele
top_5_teams_away = get_top_5_teams_home_away(avg_goals_home_away, home=False)

print("\nTop 5 Teams für Auswärtsspiele:")
for idx, team_stats in enumerate(top_5_teams_away, start=1):
    print(f"{idx}. Team: {team_stats['_id']}, Durchschnittliche Tore: {team_stats['away_goals']:.2f}")
```

Nun wollen wir uns ansehen, ob es eine Halbzeit gibt in der mehr Tore fallen, als in der anderen. Das machen wir mit der Funktion "calculate_avg_goals_per_half", welche uns ein Kreisdiagramm ausgibt mit den Prozenten. Wir können sehen, dass in dieser Champions League in beiden Hälften gleich viele Tore gefallen sind.

```python

# Funktion zum Berechnen der durchschnittlichen Anzahl der Tore pro Spielphase
def calculate_avg_goals_per_half(collection_name, half):
    collection = db[collection_name]
    pipeline = [
        {"$unwind": "$matches"},
        {"$match": {"matches.status": "FINISHED"}},
        {"$project": {
            "home_goals": "$matches.score.fullTime.home",
            "away_goals": "$matches.score.fullTime.away",
            "half": half,
        }},
        {"$group": {
            "_id": "$half",
            "avg_goals": {"$avg": {"$add": ["$home_goals", "$away_goals"]}}
        }}
    ]
    result = list(collection.aggregate(pipeline))
    return result

# Durchschnittliche Anzahl der Tore in der ersten Halbzeit berechnen
avg_goals_first_half = calculate_avg_goals_per_half('cl_matches', half='firstHalf')

# Durchschnittliche Anzahl der Tore in der zweiten Halbzeit berechnen
avg_goals_second_half = calculate_avg_goals_per_half('cl_matches', half='secondHalf')

# Daten für die Visualisierung vorbereiten
half_labels = ['First Half', 'Second Half']
avg_goals = [avg_goals_first_half[0]['avg_goals'], avg_goals_second_half[0]['avg_goals']]

# Visualisierung erstellen
plt.figure(figsize=(8, 6))
plt.pie(avg_goals, labels=half_labels, autopct='%1.1f%%', colors=['blue', 'red'], startangle=90)
plt.title('Prozentualer Anteil der durchschnittlichen Tore pro Spielphase')
plt.show()
```

Wir schauen uns jetzt dazu mal die Top 5 Torjäger in der ersten und zweiten Halbzeit an. Es fällt direkt auf, dass in der ersten Halbzeit bei den Top 5 Mannschaften im Durschnitt doppelt so viele Tore gefallen sind wie bei den Top 5 der zweiten Halbzeit.

```python

# Funktion zum Abrufen der Top-5-Teams mit den meisten Toren in einer bestimmten Halbzeit
def get_top_5_teams(collection_name, half):
    collection = db[collection_name]
    pipeline = [
        {"$unwind": "$matches"},
        {"$match": {"matches.status": "FINISHED"}},
        {"$project": {
            "home_team": "$matches.homeTeam.name",
            "away_team": "$matches.awayTeam.name",
            "home_goals": "$matches.score.fullTime.home",
            "away_goals": "$matches.score.fullTime.away",
            "half": half,
        }},
        {"$group": {
            "_id": "$home_team",
            "total_goals": {"$sum": "$home_goals" if half == 'firstHalf' else "$away_goals"}
        }},
        {"$sort": {"total_goals": -1}},
        {"$limit": 5}
    ]
    result = list(collection.aggregate(pipeline))
    return result

# Top 5 Teams mit den meisten Toren in der ersten Halbzeit abrufen
top_5_teams_first_half = get_top_5_teams('cl_matches', half='firstHalf')

# Top 5 Teams mit den meisten Toren in der zweiten Halbzeit abrufen
top_5_teams_second_half = get_top_5_teams('cl_matches', half='secondHalf')

# Daten für die Visualisierung vorbereiten
teams_first_half = [team['_id'] for team in top_5_teams_first_half]
goals_first_half = [team['total_goals'] for team in top_5_teams_first_half]

teams_second_half = [team['_id'] for team in top_5_teams_second_half]
goals_second_half = [team['total_goals'] for team in top_5_teams_second_half]

# Visualisierung erstellen
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# Erste Halbzeit
ax1.barh(teams_first_half, goals_first_half, color='blue')
ax1.set_xlabel('Anzahl der Tore')
ax1.set_ylabel('Team')
ax1.set_title('Top 5 Teams mit den meisten Toren in der ersten Halbzeit')

# Zweite Halbzeit
ax2.barh(teams_second_half, goals_second_half, color='red')
ax2.set_xlabel('Anzahl der Tore')
ax2.set_ylabel('Team')
ax2.set_title('Top 5 Teams mit den meisten Toren in der zweiten Halbzeit')

plt.tight_layout()
plt.show()
```

Schließlich identifizieren wir die Teams, die die wenigsten Gegentore pro Spiel in der Champions League erhalten haben. Diese Analyse ermöglicht es uns, die defensiven Fähigkeiten der Teams zu bewerten und zu vergleichen.


Mit der Funktion "top_5_teams_least_conceded" schauen wir uns nun auch an welche Teams, die wenigsten Gegentore bekommen haben. Da findet man zwei Teams welche mit 0,25 die wenigsten Gegentore bekommen haben.

```python

# Funktion zur Berechnung der Top 5 Teams mit den wenigsten Gegentoren pro Spiel
def top_5_teams_least_conceded(collection_name):
    collection = db[collection_name]
    pipeline = 
        {"$unwind": "$matches"},
        {"$match": {"matches.status": "FINISHED"}},
        {"$project": {
            "home_team": "$matches.homeTeam.name",
            "away_team": "$matches.awayTeam.name",
            "home_goals": "$matches.score.fullTime.home",
            "away_goals": "$matches.score.fullTime.away"
        }},
        {"$group": {
            "_id": "$home_team",
            "total_goals_conceded": {"$sum": "$away_goals"},
            "total_matches": {"$sum": 1}
        }},
        {"$project": {
            "_id": 0,
            "team": "$_id",
            "goals_conceded_per_game": {"$divide": ["$total_goals_conceded", "$total_matches"]}
        }},
        {"$sort": {"goals_conceded_per_game": 1}},
        {"$limit": 5}
    ]
    result = list(collection.aggregate(pipeline))
    return result

# Top 5 Teams mit den wenigsten Gegentoren pro Spiel für die UEFA Champions League finden
top_5_teams = top_5_teams_least_conceded('cl_matches')

# Ausgabe der Ergebnisse
for idx, team_stats in enumerate(top_5_teams, start=1):
    print(f"Platz {idx}: {team_stats['team']} - Durchschnittliche Gegentore pro Spiel: {team_stats['goals_conceded_per_game']:.2f}")


# Daten für das Diagramm vorbereiten
teams = [team['team'] for team in top_5_teams]
goals_conceded = [team['goals_conceded_per_game'] for team in top_5_teams]

# Balkendiagramm erstellen
plt.figure(figsize=(10, 6))
plt.bar(teams, goals_conceded, color='skyblue')
plt.xlabel('Team')
plt.ylabel('Durchschnittliche Gegentore pro Spiel')
plt.title('Top 5 Teams mit den wenigsten Gegentoren pro Spiel')
plt.xticks(rotation=45, ha='right')  # Rotiere die Teamnamen um 45 Grad für bessere Lesbarkeit
plt.tight_layout()  # Automatische Anpassung der Layout-Abstände
plt.show()
```
